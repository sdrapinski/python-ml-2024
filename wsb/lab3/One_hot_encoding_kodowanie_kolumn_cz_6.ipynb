###One-hot encoding - kodowanie kolumn kategorycznych

###**Import bibliotek**

```python
import pandas as pd
```

```python
df = pd.DataFrame({"x": list(range(100)),"y":["a", "b", "c", "d"]*25})
```

Przy pomocy biblioteki **one-hot encoding** będziemy chcieli zakodować kalumnę y.

```python
df.head()
```

Dla zmiennej y - zmiennej typu string utworzone zostały nowe kolumny.

Każda z tych kolumn ma taką nazwę jak nazwa pierwotnej kolumny prefiks nazwa kategorii, która tutaj występuje.

Wartościami tych zmiennych są 1 i 0.


```python
pd.get_dummies(df["y"], dtype='int').head()
```
